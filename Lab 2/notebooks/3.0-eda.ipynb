**Question:** Does `compulsory education duration` have an positive impact on average income of a citizen?

**Answering this question will:** If we know that longer compussory education duration can help develop the country, we would suggest some countries with short `compulsory education duration` to be increased lead to more effective education policies, economic development, poverty reduction, ultimately benefiting the well-being and prosperity of a nation.

**How we answer this question:** By considering how do the `GDP per Capita` (`GDP` divided by its `total population`) change when _average_ `compulsory education duration` increase? _We use this feature instead of `GDP` to remain fairness comparison between each country since they have different population._

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 

# check if all values of compulsory education dur are all equal in 2022

# get the dataframe with only the most recent year
tmp = most_recent.copy() # calculated above
# extract columns
tmp = most_recent[['Country', 'Compulsory Education Dur.', 'GDP in USD']].reset_index(drop=True)
print(tmp.head())
print(".......")
# check if all values are equal
male_check = (tmp['Compulsory Education Dur.'] == tmp['Compulsory Education Dur.'][0]).all()
female_check = (tmp['GDP in USD'] == tmp['GDP in USD'][0]).all()
print("All values in Compulsory Education Dur. are equal:", male_check)
print("All values in GDP in USD are equal:", female_check)
```

The feature `compulsory education duration` in 2022 have all same values since all of them are just a placeholder for the missing values in the original data (that we have replaced that the previous stage). However, the other feature don't belong to this case, we can also use the data in the year 2022 since we need to examine how income change when compulsory education duration increase.

Note that I'll exclude the year represent for missing value to ensure correctness. The value that represent for missing values are mean values which are not integer numbers in this case (can easily detected). However, there's also the case where mean values are integer numbers, we'll need another method in this case.

```python
tmp_data = cleaned_df.copy() # copy to avoid changes to original data

# calculate new feature: GDP per capita
tmp_data['GDP per Capita'] = tmp_data['GDP in USD'] / tmp_data['Total Population']

# extract columns
tmp_data = tmp_data[['Country', 'Year', 'GDP per Capita', 'Compulsory Education Dur.']] 

grouped = tmp_data.groupby('Compulsory Education Dur.')
# Create a new dataframe
tmp_df = pd.DataFrame()
tmp_df['Avg.GDP per Capita'] = grouped['GDP per Capita'].mean()

CEdur_by_country = [] # compulsory education duration by country
for country, durations in tmp_data.groupby('Country')['Compulsory Education Dur.']:
    dur = []
    for s in durations.unique().tolist():
        if s - int(s) == 0: # exclude mising values
            dur.insert(0, int(s))
    CEdur_by_country.append((country, dur))

maxlen = -1
for item in CEdur_by_country:
    if len(item[1]) > maxlen:
        maxlen = len(item[1])
for i, item in enumerate(CEdur_by_country):
    item[1][:0] = [0 for _ in range(maxlen - len(item[1]))]
CEdur_by_country.sort(key=lambda x: x[1][-1])
```

### Analyzing to get the answer

```python
# Extract countries and education durations
countries = [item[0] for item in CEdur_by_country]
durations = [item[1] for item in CEdur_by_country]

# Create the stacked bar chart
plt.figure(figsize=(12, 8))
bottom = [0] * len(countries)

for i in range(maxlen - 1, -1, -1):
    plt.bar(countries, [dur[i] for dur in durations], label=f'Duration {i + 1}')

plt.xlabel('Countries')
plt.ylabel('Compulsory Education Duration (years)')
plt.title('Compulsory Education Duration by Country')
plt.xticks(rotation=90)
plt.legend(title='Duration')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# YOUR CODE HERE
fig, ax = plt.subplots(1, 1)
ax = sns.lineplot(x=tmp_df.index, y=tmp_df['Avg.GDP per Capita'])
plt.xlabel('Compulsory Education Duration')
plt.ylabel('Average GDP per Capita (in USD)')
plt.show()
```

__Observation:__
